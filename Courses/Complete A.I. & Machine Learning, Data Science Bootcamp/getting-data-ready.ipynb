## Getting our data ready

Three main things we have to do:

    1. Split the data into features and labels 
    2. Filling (also called imputing) or disregarding missing values
    3. Converting non-numerical values to numerical values (feature encoding)

```python
heart_disease = pd.read_csv("data/heart-disease.csv")
```

```python
heart_disease.head()
```

### Split the data into features and labels

```python
X = heart_disease.drop("target", axis=1)
X.head()
```

```python
y = heart_disease["target"]
y.head()
```

### Split the data into training and test sets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
car_sales = pd.read_csv("data/car-sales-extended.csv")
```

```python
car_sales.head()
```

```python
# Split into X and y
X = car_sales.drop("Price", axis=1)
y = car_sales["Price"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Turn the categories into numbers
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

categorical_features = ["Make", "Colour", "Doors"]
one_hot = OneHotEncoder()
transformer = ColumnTransformer([
    ("one_hot",
    one_hot,
    categorical_features)],
    remainder="passthrough")
transformed_X = transformer.fit_transform(X)
transformed_X
```

```python
pd.DataFrame(transformed_X)
```

```python
dummies = pd.get_dummies(car_sales[["Make", "Colour", "Doors"]], dtype=int)
dummies
```

```python
# Refitting the model
from sklearn.ensemble import RandomForestRegressor
np.random.seed(42)
X_train, X_test, y_train, y_test = train_test_split(transformed_X, y, test_size=0.2)

model = RandomForestRegressor()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

## Choosing the right estimator/algorithm

Some things to note
* Sci-kit learn refers to ml models, algorithms as estimators
* Classification problem - predict a category
* Regression problem - predicting a number

https://scikit-learn.org/1.3/tutorial/machine_learning_map/index.html

```python
# Using california housing dataset
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
housing
```

```python
housing_df = pd.DataFrame(housing["data"], columns=housing["feature_names"])
housing_df
```

```python
housing_df['target'] = housing["target"]
housing_df = housing_df.drop("MedHouseVal", axis=1)
```

```python
from sklearn.linear_model import Ridge

np.random.seed(40)

X = housing_df.drop("target", axis=1)
y = housing_df["target"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model = Ridge()
model.fit(X_train, y_train)

model.score(X_test, y_test)
```
